### 통계학 강의 1

- **통계학 정의** <br>
수학에서 출발한 가지로, 데이터를 모으고 분석하고 해석하고 나타내는 학문이다. 주로 산업에서 혹은 사회적인 문제를 다룰 때 많이 사용된다. 크게 두 가지 기술통계(descriptive statistics)과 추론통계(inference statistics)로 나뉜다.


- **기술 통계** <br>
측정, 실험에서 수집한 자료를 정리, 표현, 요약, 해석해서 자료의 특성을 규명하는 내용을 다루는 학문이다. 주로 분산, 평균, 표준편차 등이 많이 쓰인다.  



- **추론 통계** <br>
수집한 자료를 바탕으로 의미있는 결론을 추론하기 위한 학문이다. 자료를 모집단 전부들 대상으로 수집하는 것은 현실적으로 불가능하며 많은 비용이 드는 일이라 샘플링된 자료를 바탕으로 모집단을 예측하는 일이 필요하다. (예를 들면, 출구 조사를 토대로 대통령 당선 결과 맞추기.) 
추론하려는 항목들도 다양한데 yes/no 질문(hypothesis testing), 추정(estimation), 상관관계(correlation), 모델링(회귀 분석) 등이 있다. 이 분야가 머신러닝 분야랑 깊게 관련이 있으며 현재에는 둘 사이에 구분이 거의 없다. <br>
(출처 : [위키피디아](https://en.wikipedia.org/wiki/Statistics))


- **통계학에 대한 나의 생각** <br>
기초통계학은 누구나 배워야하는 학문이라고 생각한다. 왜냐하면 우리가 정보를 보는 눈을 길러주기 때문이다. <br>
그럼 먼저, 이런 눈을 길러서 어디에 쓸까? 크게는 어떤 선택해야 하는 상황에서 조금은 더 유리한 선택을 할 수 있고 잘못된 정보로부터 오는 손해도 피할 수 있다. <br>
다음으로는, 왜 편향된 정보를 우리에게 주는 것일까? 당연히 이익을 얻기 위해서이다. 우리는 매일 매체, 정부, 기관, 기업들로부터 많은 ***편향된 정보*** 를 받는다. 
*(통계학에서는 bias(편향된)이라는 단어를 많이 쓰는데 한 쪽으로 치우쳐져 있다는 표현이다.)*  <br>
모든 것들을 이런 방식으로 생각하기는 힘들겠지만 어떤 행동에는 항상 목적이 있고 이유가 있다. 편향된 정보를 굳이 만들었다고 하면 그 나름의 이유가 있을 것이고 우리가 그렇게 생각하기를 바라서가 아닐까? (물론 우연도 있다.)
가장 쉬운 예로 우리가 **이슬람, 흑인**에 대해 가지고 있는 생각들이라던지 **국산품**에 대해 가지고 있는 생각들이 그런 것이지 않을까 싶다. <br>
**이슬람**의 경우 궁금하지도 않고 접해본 적도 없지만 어떤 이미지를 가지고 있지 않나? 
또, **국산품**에 대해서 어떤 좋은 이미지가 있나? 진짜 해외 제품보다 좋으면 왜 해외 시장에 못 나갈까? <br>
만약 특정 대상에 대해 어떤 이미지를 가지고 있다면 어디에서 온 것인지 생각해볼 필요가 있다. 어느 집단이 내가 그렇게 생각하길 바라는 지도 생각해보자. 그리고 그런 이미지를 아무 이유 없이 갖고 싶은지. <br>
나는 싫다. 그래서 특정 주제에 대해서 판단을 하기 위해서는 최소한의 자료를 조사해야 하고 나의 생각에 대한 Yes/No 결론을 내리기 위해서는 타당한 근거와 **어느 정도로 Yes/No 인지** 알 수 있어야 한다고 생각한다. 그리고 이런 생각을 할 수 있게 통계학이 도와준다고 생각해 많은 사람들이 통계학을 배우면 편향된 시각에서 벗어나는데 도움이 될 것이라 생각한다..<br>
